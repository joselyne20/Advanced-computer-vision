```
import os
import cv2

# Define base directory for your dataset (replace with your actual path)
BASE_DIR = '/content/drive/MyDrive/sem 2/advanced cv/FashionDataset'  # Assuming your data is mounted on Colab

# Define subdirectories containing data (adjust as needed)
IMAGES_DIR = os.path.join(BASE_DIR, 'img')  # Assuming images are in a subdirectory 'images'
ANNOTATIONS_DIR = os.path.join(BASE_DIR, 'split')
TRAIN_IMAGES_FILE = os.path.join(ANNOTATIONS_DIR, 'train.txt')
TRAIN_LABELS_FILE = os.path.join(ANNOTATIONS_DIR, 'train_attr.txt')  # Assuming label file name

def match_and_display_images(image_paths_file, labels_file, num_images=10):
  """
  Matches image paths in a file with labels in another file,
  assuming a one-to-one correspondence based on line order.
  Prints the image path, label, and displays the image using OpenCV.

  Args:
      image_paths_file (str): Path to the file containing image paths.
      labels_file (str): Path to the file containing labels.
      num_images (int, optional): Number of image-label pairs to print and display. Defaults to 10.
  """
  # Check if OpenCV is available
  try:
    cv2.waitKey(1)  # Basic check for OpenCV availability
  except Exception as e:
    print("Error: OpenCV is not installed. Please install OpenCV to display images.")
    return

  with open(image_paths_file, 'r') as f_images, open(labels_file, 'r') as f_labels:
    image_paths = [line.strip() for line in f_images]
    labels = [line.strip() for line in f_labels]

    # Check if number of image paths and labels match
    if len(image_paths) != len(labels):
      print(f"Error: Number of image paths ({len(image_paths)}) doesn't match number of labels ({len(labels)}).")
      return

    # Print and display the first 'num_images' image-label pairs
    for i in range(min(num_images, len(image_paths))):
      image_path = image_paths[i]
      label = labels[i]

      # Construct the complete image path
      complete_image_path = os.path.join(IMAGES_DIR, image_path)  # Assuming images are in 'images' subdirectory

      # Read the image
      try:
        image = cv2.imread(complete_image_path)
        if image is None:
          print(f"Warning: Image not found - {complete_image_path}")
          continue
      except Exception as e:
        print(f"Error reading image: {complete_image_path} ({e})")
        continue

      # Display the image with label
      cv2.imshow('Image', image)
      cv2.waitKey(0)  # Wait for a key press to close the window
      cv2.destroyWindow('Image')

      print(f"Image path: {complete_image_path}")
      print(f"Label: {label}")
      print("-" * 20)  # Optional separator between entries

  print(f"Printed and displayed {min(num_images, len(image_paths))} image-label pairs.")


if __name__ == "__main__":
  match_and_display_images(TRAIN_IMAGES_FILE, TRAIN_LABELS_FILE)
```
